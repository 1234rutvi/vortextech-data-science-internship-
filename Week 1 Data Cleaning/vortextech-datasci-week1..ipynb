# Week 1: Data Cleaning and Basic Visualization
**Dataset:** Super Store

## Objective
Clean the dataset and create meaningful visualizations.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.rcParams['figure.figsize']=(10,6)
df=pd.read_csv('/content/super store(1).csv')
df.head()
```

## Dataset Overview

```
print(df.shape)
display(df.head())
display(df.info())
display(df.describe(include='all'))
```

## Check Missing Values

```
missing=df.isnull().sum()
display(missing)
```

No significant missing values should be handled if present.

## Duplicate Records

```
print('Duplicates:',df.duplicated().sum())
df.drop_duplicates(inplace=True)
print('Shape after removing duplicates:',df.shape)
```

## Data Types

```
display(df.dtypes)
```

```
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df['Order Month']=df['Order Date'].dt.to_period('M').astype(str)
df['Shipping Days']=(df['Ship Date']-df['Order Date']).dt.days
df.head()
```

## Category

```
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Category',y='Sales',estimator=np.sum)
plt.title('Category')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Technology/Furniture/Office Supplies comparison.

## Region

```
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Region',y='Sales',estimator=np.sum)
plt.title('Region')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Regional sales comparison.

## Segment

```
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Segment',y='Sales',estimator=np.sum)
plt.title('Segment')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Customer segment contribution.

## Ship Mode

```
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='Ship Mode')
plt.title('Ship Mode')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Order count by shipping mode.

## Sales Distribution

```
plt.figure(figsize=(10,5))
plt.hist(df['Sales'],bins=30)
plt.title('Sales Distribution')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Distribution of sales values.

## Monthly Sales

```
plt.figure(figsize=(10,5))
df.groupby('Order Month')['Sales'].sum().plot(marker='o')
plt.title('Monthly Sales')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**Insight:** Monthly trend.

## Summary of Findings
- Dataset cleaned successfully.
- Duplicate records removed.
- Dates converted into datetime.
- Multiple visualizations reveal category, regional, customer segment and shipping patterns.
- Dataset is now ready for advanced exploratory analysis (Week 2).
